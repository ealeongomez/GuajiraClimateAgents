```python
import os
import pymssql
from datetime import datetime
from dotenv import load_dotenv

load_dotenv()
```

```python
conn = pymssql.connect(
    server=os.getenv("DB_SERVER", "localhost"),
    user=os.getenv("DB_USER", "sa"),
    password=os.getenv("DB_PASSWORD"),
    database=os.getenv("DB_NAME", "ClimateDB"),
    port=os.getenv("DB_PORT", "1433")
)

cursor = conn.cursor(as_dict=True)
```

```python
query = """
SELECT 
    MAX(datetime) as last_update,
    DATEDIFF(HOUR, MAX(datetime), GETDATE()) as hours_behind,
    DATEDIFF(MINUTE, MAX(datetime), GETDATE()) as minutes_behind
FROM climate_observations;
"""
```

```python
cursor.execute(query)
result = cursor.fetchone()

print("\n" + "=" * 60)
print("üìä ESTADO DE LA BASE DE DATOS")
print("=" * 60)
print(f"‚è∞ Hora actual: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"üìÖ √öltima actualizaci√≥n: {result['last_update']}")
print(f"‚è±Ô∏è  Atraso: {result['hours_behind']} horas, {result['minutes_behind'] % 60} minutos")

if result['hours_behind'] <= 1:
    print("üéØ Estado: ‚úÖ ACTUALIZADO")
elif result['hours_behind'] <= 24:
    print("üéØ Estado: ‚ö†Ô∏è ATRASADO")
else:
    print("üéØ Estado: ‚ùå MUY ATRASADO")

print("=" * 60 + "\n")

cursor.close()
conn.close()
```
